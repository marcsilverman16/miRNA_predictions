# Plot Statistics for Data

```python
# plot stats (script from dataset_stats.py updated with proper directory)
import numpy as np
import pandas as pd
import os,random
from sklearn.manifold import TSNE
from matplotlib import pyplot as plt
from sklearn.ensemble import RandomForestClassifier
import pickle
data_dir="/home/hice1/slinen3/scratch/miRNA_data/features/"
gen_names=["A. thaliana","C. elegans","A. gambiae","H. sapiens","D. melanogaster"]
genomes=["ath","cel","aga","hsa","dme"]

random.seed(3)
np.random.seed(3)

ranks=[]
all_pos_samples=[]
all_neg_samples=[]

for gen_name,genome in zip(gen_names,genomes):

    print(genome)
    if genome=="hsa": # it is a huge file, using a smaller subsample
        try:
            df=pd.read_csv(data_dir+"hsa_miRNAs.csv")
        except:
            print("Please unzip features to use this script")
            exit()
        df=df.drop(['sequence_names'],axis=1)
        pos_samples=df.values.astype(np.float)

        neg_samples=pickle.load(open(data_dir+"hsa_unlabeled_subsample.pk","rb"))[:,1:].astype(np.float)
        reduced_data=np.concatenate((pos_samples,neg_samples),axis=0)
        reduced_labels=np.concatenate((np.ones(pos_samples.shape[0]),np.zeros(neg_samples.shape[0])),axis=0)

    else:
        df=pd.read_csv(data_dir+genome+".csv",low_memory=False)
        df=df.drop(['sequence_names'],axis=1)
        pos_samples=df[df.CLASS==1].values.astype(float)
        neg_samples=df[df.CLASS==0].sample(10000).values.astype(float)
        reduced_data=np.concatenate((pos_samples[:,:-1],neg_samples[:,:-1]),axis=0)
        reduced_labels=np.concatenate((pos_samples[:,-1],neg_samples[:,-1]),axis=0)

    # zscore
    # Feature Normalization ==========
    for f in range(reduced_data.shape[1]):
        m=np.mean(reduced_data[:,f])
        s=np.std(reduced_data[:,f])

        reduced_data[:,f]=reduced_data[:,f]-m
        reduced_data[:,f]=reduced_data[:,f]/s

        reduced_data[np.where(np.isnan(reduced_data[:,f]))[0],f]=0
        reduced_data[np.where(np.isinf(reduced_data[:,f]))[0],f]=0

    pos=reduced_data[reduced_labels==1,:]
    neg=reduced_data[reduced_labels==0,:]
    all_pos_samples.append(pos)
    all_neg_samples.append(neg)

    #t-SNE
    data_tsne=TSNE().fit_transform(reduced_data)
    npos=pos_samples.shape[0]
    s=5
    alpha=.8
    plt.figure()
    plt.scatter(data_tsne[npos:,0],data_tsne[npos:,1],s=s,alpha=alpha,label="Unlabeled")
    plt.scatter(data_tsne[:npos,0],data_tsne[:npos,1],s=s*3,alpha=alpha,label="Positives")
    plt.legend()
    plt.title(gen_name)

    ##  Feature importance using random forest classifiers.
    forest = RandomForestClassifier(n_estimators=10)

    forest.fit(reduced_data, reduced_labels)
    importances = forest.feature_importances_
    indices = np.argsort(importances)[::-1]

    ranks.append(list(indices))


# Show the best 5 features in average rank
M=reduced_data.shape[1]
avrank=np.zeros((M))
for f in range(M):
    avrank[f]=np.mean([r.index(f) for r in ranks])


avrank_ind=np.argsort(avrank)
for f in range(M):
    print("%s: %.2f" %(df.columns[avrank_ind[f]],avrank[avrank_ind[f]]))


# plot histograms
for t,pos,neg in zip(gen_names,all_pos_samples,all_neg_samples):
    plt.figure(t)
    k=1
    for f in avrank_ind[:3]:
        plt.subplot(2,3,k)
        plt.hist(pos[:,f],bins=100,color="C1")
        plt.xlim([-10,10])
        plt.subplot(2,3,k+3)
        plt.hist(neg[:,f],bins=100,color="C0")
        plt.xlabel("Feature normalized value")
        plt.ylabel("Number of samples")
        plt.title(df.columns[f])
        plt.xlim([-10,10])

        k+=1
plt.show()
```

```python
# plot stats (script from dataset_stats.py updated with proper directory)
import numpy as np
import pandas as pd
import os, random
from sklearn.manifold import TSNE
from matplotlib import pyplot as plt
from sklearn.ensemble import RandomForestClassifier
import pickle

data_dir = "/home/hice1/slinen3/scratch/miRNA_data/features/"
gen_names = ["A. thaliana", "C. elegans", "A. gambiae", "H. sapiens", "D. melanogaster"]
genomes = ["ath", "cel", "aga", "hsa", "dme"]

random.seed(3)
np.random.seed(3)

ranks = []
all_pos_samples = []
all_neg_samples = []

for gen_name, genome in zip(gen_names, genomes):
    print(genome)
    if genome == "hsa":  # it is a huge file, using a smaller subsample
        try:
            df = pd.read_csv(data_dir + "hsa_miRNAs.csv")
        except:
            print("Please unzip features to use this script")
            exit()
        df = df.drop(['sequence_names'], axis=1)
        pos_samples = df.values.astype(np.float)
        neg_samples = pickle.load(open(data_dir + "hsa_unlabeled_subsample.pk", "rb"))[:, 1:].astype(np.float)
        reduced_data = np.concatenate((pos_samples, neg_samples), axis=0)
        reduced_labels = np.concatenate((np.ones(pos_samples.shape[0]), np.zeros(neg_samples.shape[0])), axis=0)
    else:
        df = pd.read_csv(data_dir + genome + ".csv", low_memory=False)
        df = df.drop(['sequence_names'], axis=1)
        pos_samples = df[df.CLASS == 1].values.astype(float)
        neg_samples = df[df.CLASS == 0].sample(10000).values.astype(float)
        reduced_data = np.concatenate((pos_samples[:, :-1], neg_samples[:, :-1]), axis=0)
        reduced_labels = np.concatenate((pos_samples[:, -1], neg_samples[:, -1]), axis=0)

    # Feature Normalization
    for f in range(reduced_data.shape[1]):
        m = np.mean(reduced_data[:, f])
        s = np.std(reduced_data[:, f])

        reduced_data[:, f] = (reduced_data[:, f] - m) / s
        reduced_data[np.isnan(reduced_data[:, f]), f] = 0
        reduced_data[np.isinf(reduced_data[:, f]), f] = 0

    pos = reduced_data[reduced_labels == 1, :]
    neg = reduced_data[reduced_labels == 0, :]
    all_pos_samples.append(pos)
    all_neg_samples.append(neg)

    # t-SNE
    data_tsne = TSNE().fit_transform(reduced_data)
    npos = pos_samples.shape[0]
    s = 5
    alpha = .8
    plt.figure()
    plt.scatter(data_tsne[npos:, 0], data_tsne[npos:, 1], s=s, alpha=alpha, label="Unlabeled")
    plt.scatter(data_tsne[:npos, 0], data_tsne[:npos, 1], s=s * 3, alpha=alpha, label="Positives")
    plt.legend()
    plt.title(gen_name)

    # Save the t-SNE plot
    plt.savefig('tsne_plot_{}.png'.format(genome))

    # Feature importance using random forest classifiers.
    forest = RandomForestClassifier(n_estimators=10)
    forest.fit(reduced_data, reduced_labels)
    importances = forest.feature_importances_
    indices = np.argsort(importances)[::-1]
    ranks.append(list(indices))

# Show the best 5 features in average rank
M = reduced_data.shape[1]
avrank = np.zeros((M))
for f in range(M):
    avrank[f] = np.mean([r.index(f) for r in ranks])

avrank_ind = np.argsort(avrank)
for f in range(M):
    print("%s: %.2f" % (df.columns[avrank_ind[f]], avrank[avrank_ind[f]]))

# plot histograms
for t, pos, neg in zip(gen_names, all_pos_samples, all_neg_samples):
    plt.figure(t)
    k = 1
    for f in avrank_ind[:3]:
        plt.subplot(2, 3, k)
        plt.hist(pos[:, f], bins=100, color="C1")
        plt.xlim([-10, 10])
        plt.subplot(2, 3, k + 3)
        plt.hist(neg[:, f], bins=100, color="C0")
        plt.xlabel("Feature normalized value")
        plt.ylabel("Number of samples")
        plt.title(df.columns[f])
        plt.xlim([-10, 10])
        k += 1

    # Save the histograms plot
    plt.savefig('histograms_{}.png'.format(t))

plt.show()
```
